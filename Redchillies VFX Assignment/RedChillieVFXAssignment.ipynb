```python
import time
import psutil
import matplotlib.pyplot as plt
import cpuinfo
```

```python
%matplotlib notebook
plt.rcParams['animation.html']='jshtml'
```

```python
print("frequency of CPU ==>> ",psutil.cpu_freq(percpu=True))
print("No. of CPU available ==>>",psutil.cpu_count(logical=False))
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
a=cpuinfo.get_cpu_info()["brand"]                #this will select cpu information and brand name from a dictionary
plt.title(a)                                     #this will print cpu information and brand name
plt.xlabel('Time')                               #x-axis value
plt.ylabel('captured event data')                #y-axis value
fig.show()
```

```python
i=0
x,y=[],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y, color='b')
    ax.set_xlim(left=max(0,i-50),right=i+50)
    fig.canvas.draw()
    time.sleep(0.1)
    i+=1
```
